## 数据转换

```python
def generate_index_map(lables):
    index_label = {}
    label_index = {}
    i = 0
    for l in lables:
        index_label[i] = l
        label_index[l] = i
        i += 1
    return label_index, index_label
```

```python
states_label_index, states_index_label = generate_index_map(states)
observations_label_index, observations_index_label = generate_index_map(observations)
```

```python
def convert_observations_to_index(observations, label_index):
    list = []
    for o in observations:
        list.append(label_index[o])
    return list
```

```python
def convert_map_to_vector(map, label_index):
    v = np.empty(len(map), dtype=float)
    for e in map:
        v[label_index[e]] = map[e]
    return v
```

```python
def convert_map_to_matrix(map, label_index1, label_index2):
    m = np.empty((len(label_index1), len(label_index2)), dtype=float)
    for line in map:
        for col in map[line]:
            m[label_index1[line]][label_index2[col]] = map[line][col]
    return m
```

```python
A = convert_map_to_matrix(transition_probability, states_label_index, states_label_index)
B = convert_map_to_matrix(emission_probability, states_label_index, observations_label_index)
observations_index = convert_observations_to_index(observations, observations_label_index)
pi = convert_map_to_vector(start_probability, states_label_index)
```

### 根据发射概率矩阵采样观察

```python
def generate_observation(emission_probability, current_state):
    rd = np.random.rand()
    value1 = emission_probability[current_state][observations_index_label[0]]
    value2 = emission_probability[current_state][observations_index_label[1]]
    if rd <= value1:
        return 0
    elif rd > value1 and rd <= value1 + value2:
        return 1
    else:
        return 2
    
observations_index_label[generate_observation(emission_probability, "Healthy")]
```

### 生成序列

```python
def generate(length):
    hidden_states = []
    observations = []
    current_state = states_index_label[generate_initial_state(start_probability)]
    hidden_states.append(current_state)
    observations.append(observations_index_label[generate_observation(emission_probability, current_state)])
    for i in range(1, length):
        current_state = states_index_label[generate_transition_state(transition_probability, current_state)]
        hidden_states.append(current_state)
        observations.append(observations_index_label[generate_observation(emission_probability, current_state)])
    return hidden_states,observations

states,observations = generate(5)
```

## viterbi 算法

```python
def viterbi(trainsition_probability,emission_probability,pi,obs_seq):

    # 隐藏状态个数
    N = np.array(trainsition_probability).shape[0] 
    
    # 观测序列的观测个数，即时刻个数
    T = len(obs_seq) 
    
    # 每个时刻每个状态对应的局部最优状态序列的概率值
    delta = np.zeros((N,T))
    
    # 保存每个时刻每个状态取到最大概率的前置节点
    phi = np.zeros((N,T),dtype = int)
    
    #初始状态
    delta[:,0] = pi*np.transpose(emission_probability[:,obs_seq[0]])
    print("delta[:,0]",delta[:,0])

    for t in range(1,T):
        list_max=[]
        for n in range(N):
            # 计算时刻t,状态为n的所有单个路径的概率值
            delta_ti = delta[:,t-1]*np.transpose(trainsition_probability[:,n])
            print("delta_ti",delta_ti)
            # 保存最大概率
            list_max.append(np.max(delta_ti))
            print("list_max",list_max)
            # 保存最大概率对应的前置节点
            phi[n,t] = np.argmax(delta_ti)
            print("phi[%d, %d]"%(n,t), phi[n,t])    
        delta[:,t] = np.array(list_max)*np.transpose(emission_probability[:,obs_seq[t]]) 
        print("delta[:,%d]"%t,delta[:,t])

    result = np.zeros(T)
    result[T-1] = np.argmax(delta[:,T-1])
    print("result[%d]"%(T-1), result[T-1])
    for t in range(T-2,-1,-1):
        result[t] = phi[int(result[t+1]),t+1]
        print("result[%d]"%t, result[t])
    return result
```

```python
def viterbi_with_flowchart(trainsition_probability,emission_probability,pi,states, observations,obs_seq):
    
    g = Digraph()
    g.attr(rankdir='LR')
    
    # 隐藏状态个数
    N = np.array(trainsition_probability).shape[0] 
    
    # 观测序列的观测个数，即时刻个数
    T = len(obs_seq) 
    
    # 每个时刻每个状态对应的局部最优状态序列的概率值
    delta = np.zeros((N,T))
    
    # 保存每个时刻每个状态取到最大概率的前置节点
    phi = np.zeros((N,T),dtype = int)
    
    #初始状态
    delta[:,0] = pi*np.transpose(emission_probability[:,obs_seq[0]])
    logging.debug("delta[:,0]: " + str(delta[:,0]))
    
    for index,state in enumerate(states):
        g.node(str(0) + "_" + state, label=state)
        g.edge("Start", str(0) + "_" + state)
    
    for t in range(1,T):
        logging.debug("\t 第一层for循环：t = %d"%t)
        for index,state in enumerate(states):
            g.node(str(t) + "_" + state, label=state)
        list_max=[]
        for n in range(N):
            logging.debug("\t\t 第二层for循环：n = %d"%n)
            # 计算时刻t,状态为n的所有单个路径的概率值
            delta_ti = delta[:,t-1]*np.transpose(trainsition_probability[:,n])
            logging.debug("\t\tdelta_ti: "+ str(delta_ti))
            # 保存最大概率
            list_max.append(np.max(delta_ti))
            logging.debug("\t\tlist_max: "+ str(list_max))
            # 保存最大概率对应的前置节点
            phi[n,t] = np.argmax(delta_ti)
            logging.debug("\t\tphi[%d, %d]: "%(n,t)+str(phi[n,t]))
            g.edge(str(t-1)+"_"+states[phi[n,t]], str(t) + "_" + states[n])
        delta[:,t] = np.array(list_max)*np.transpose(emission_probability[:,obs_seq[t]]) 
        logging.debug("\tdelta[:,%d]: "%t + str(delta[:,t]))

    result = np.zeros(T)
    result[T-1] = np.argmax(delta[:,T-1]) 
    g.node(str(T-1) + "_" + states[int(result[T-1])], label=states[int(result[T-1])], style='filled', color="lightgray")
    logging.debug("result[%d]: "%(T-1)+str(result[T-1]))
    for t in range(T-2,-1,-1):
        result[t] = phi[int(result[t+1]),t+1]
        logging.debug("result[%d]: "%t+str(result[t]))
        g.node(str(t) + "_" + states[int(result[t])], label=states[int(result[t])], style='filled', color="lightgray")
    return result,g
```

```python
def viterbi(obs, states, start_p, trans_p, emit_p):
    V = [{}]  # tabular
    path = {}
    for y in states:  # init
        V[0][y] = start_p[y] + emit_p[y].get(obs[0], MIN_FLOAT)
        path[y] = [y]
    print(path)
    for t in range(1, len(obs)):
        V.append({})
        newpath = {}
        for y in states:
            em_p = emit_p[y].get(obs[t], MIN_FLOAT)
            (prob, state) = max(
                [(V[t - 1][y0] + trans_p[y0].get(y, MIN_FLOAT) + em_p, y0) for y0 in PrevStatus[y]])
            V[t][y] = prob
            newpath[y] = path[state] + [y]
        path = newpath
        print(path)

    (prob, state) = max((V[len(obs) - 1][y], y) for y in 'ES')

    return (prob, path[state])

sentence = "商品和服务"
print(viterbi(sentence, "BMES", start_p, trans_p, emit_p))
```

```python
def __cut(sentence):
    global emit_P
    prob, pos_list = viterbi(sentence, 'BMES', start_p, trans_p, emit_p)
    begin, nexti = 0, 0
    # print pos_list, sentence
    for i, char in enumerate(sentence):
        pos = pos_list[i]
        if pos == 'B':
            begin = i
        elif pos == 'E':
            yield sentence[begin:i + 1]
            nexti = i + 1
        elif pos == 'S':
            yield char
            nexti = i + 1
    if nexti < len(sentence):
        yield sentence[nexti:]
print(list(__cut(sentence)))
```

```python
def sign_sentence(sentence):
    sentence_list = sentence.split()
    sentence_BMSE_list = [''.join(word) for word in [sign_BMSE(word) for word in sentence_list]]
    return ' '.join(sentence_BMSE_list)

BMSE_dict = {1:['S'],2:['B','E']}
def sign_BMSE(word):
    wl = len(word)
    try:
        return BMSE_dict[wl]
    except:
        if wl < 1:
            return "ERROR"
        else:
            return ['B']+['M']*(wl-2)+['E']
```

```python
def GetFrequency(dict1):
    dict2 = {}
    sum1 = sum(dict1.values())
    for key,value in dict1.items():
        dict2[key]=log(value/sum1)
    return dict2
```

## 构建状态转移概率矩阵（msr_prob_trans）

```python
def getProbTransDict(dict1):
    """
    功能：由词频字典生成转移矩阵
    """
    dict2 = {}
    dict1.keys()
    for key,value in dict1.items():
        if key[0] not in dict1.keys():
            dict2.setdefault(key[0],{})
        dict2[key[0]][key[1]] = value
    for key in dict2.keys():
        dict2[key] = GetFrequency(dict2[key])
    return dict2
```

### 构建基于msr语料库的HMM模型

```python
def viterbi(obs, states, start_p, trans_p, emit_p):
    V = [{}]  # tabular
    path = {}
    for y in states:  # init
        V[0][y] = start_p[y] + emit_p[y].get(obs[0], MIN_FLOAT)
        path[y] = [y]
#     print(path)
    for t in range(1, len(obs)):
        V.append({})
        newpath = {}
        for y in states:
            em_p = emit_p[y].get(obs[t], MIN_FLOAT)
            (prob, state) = max(
                [(V[t - 1][y0] + trans_p[y0].get(y, MIN_FLOAT) + em_p, y0) for y0 in PrevStatus[y]])
            V[t][y] = prob
            newpath[y] = path[state] + [y]
        path = newpath
#         print(path)

    (prob, state) = max((V[len(obs) - 1][y], y) for y in 'ES')

    return (prob, path[state])

sentence = "徐翔是恶霸"
print(viterbi(sentence, "BMES", msr_prob_start, msr_prob_trans, msr_prob_emit))
```

```python
def load_dictionary(dict_file):
    """
    加载词库
    :return: 一个set形式的词库
    """
    fr = open(dict_file,encoding="utf-8")
    word_list = [item.strip().split("\t")[0] for item in fr]
    return set(word_list)

word_dict = load_dictionary("../../SogouW/Freq/SogouLabDic.dic")
```

```python
def to_region(segmentation: str) -> list:
    """
    将分词结果转换为区间
    :param segmentation: 商品 和 服务
    :return: [(0, 2), (2, 3), (3, 5)]
    """
    region = []
    start = 0
    for word in re.compile("\\s+").split(segmentation.strip()):
        end = start + len(word)
        region.append((start, end))
        start = end
    return region
```

```python
def prf(gold: str, pred: str, dic) -> tuple:
    """
    计算P、R、F1
    :param gold: 标准答案文件，比如“商品 和 服务”
    :param pred: 分词结果文件，比如“商品 和服 务”
    :param dic: 词典
    :return: (P, R, F1, OOV_R, IV_R)
    """
    A_size, B_size, A_cap_B_size, OOV, IV, OOV_R, IV_R = 0, 0, 0, 0, 0, 0, 0
    with open(gold,encoding="utf-8") as gd, open(pred,encoding="utf-8") as pd:
        for g, p in zip(gd, pd):
            A, B = set(to_region(g)), set(to_region(p))
            A_size += len(A)
            B_size += len(B)
            A_cap_B_size += len(A & B)
            text = re.sub("\\s+", "", g)
            for (start, end) in A:
                word = text[start: end]
                if word in dic:
                    IV += 1
                else:
                    OOV += 1

            for (start, end) in A & B:
                word = text[start: end]
                if word in dic:
                    IV_R += 1
                else:
                    OOV_R += 1
    p, r = A_cap_B_size / B_size * 100, A_cap_B_size / A_size * 100
    return p, r, 2 * p * r / (p + r), OOV_R / OOV * 100, IV_R / IV * 100

#设置评测文本
msr_output = os.path.join(sighan05, 'testing', 'msr_output.txt')
msr_gold = os.path.join(sighan05, 'gold', 'msr_test_gold.utf8')
```

#### 一元语言模型

```python
import re
re_eng = re.compile('[a-zA-Z0-9]', re.U)
def gen_pfdict(f):
    lfreq = {}
    ltotal = 0
    for lineno, line in enumerate(f, 1):
        try:
            line = line.strip()
            word, freq = line.split(' ')[:2]
            freq = int(freq)
            lfreq[word] = freq
            ltotal += freq
            for ch in range(len(word)):
                wfrag = word[:ch + 1]
                if wfrag not in lfreq:
                    lfreq[wfrag] = 0 # 不在词典里的前缀，词频设为零，前提是词典已经进行排序
        except ValueError:
            raise ValueError(
                'invalid dictionary entry in %s at Line %s: %s' % (f.name, lineno, line))
    f.close()
    return lfreq, ltotal
f = open("../../jieba_dict.txt",encoding="utf-8")
FREQ, total = gen_pfdict(f)
def get_DAG(sentence, f):
    DAG = {}
    N = len(sentence)
    for k in range(N):
        tmplist = []
        i = k
        frag = sentence[k]
        while i < N and frag in FREQ:
            if FREQ[frag]:
                tmplist.append(i)
            i += 1
            frag = sentence[k:i + 1]
        if not tmplist:
            tmplist.append(k)
        DAG[k] = tmplist
    return DAG
def calc(sentence, DAG, route):
    N = len(sentence)
    route[N] = (0, 0)
    logtotal = log(total)
    for idx in range(N-1, -1, -1):
        tmp = []
        for x in DAG[idx]:
            prob = log(FREQ.get(sentence[idx:x + 1]) or 1) - logtotal
            value = round(prob + route[x + 1][0],2)
            tmp.append( (value, x) )
        route[idx] = max(tmp)
def __cut_DAG_NO_HMM(sentence, f):
    DAG = get_DAG(sentence, f)
    route = {}
    calc(sentence, DAG, route)
    x = 0
    N = len(sentence)
    buf = ''
    while x < N:
        y = route[x][1] + 1
        l_word = sentence[x:y]
        if re_eng.match(l_word) and len(l_word) == 1:
            buf += l_word
            x = y
        else:
            if buf:
                yield buf
                buf = ''
            yield l_word
            x = y
    if buf:
        yield buf
        buf = ''
            
list(__cut_DAG_NO_HMM(sentence,f))
```

#### 改进版后项最大匹配

```python
word_length = [len(i) for i in word_dict]#word_length内储存词典里每一个词的长度
max_word_length = max(word_length)#max_word_length最大词长

def mini_segment(text,begin,dic,max_length):
    """text：输入文本
       begin：指针开始位置
       dic：词典
       max_length：最长字符长度"""
    for i in range(max_length,0,-1):
        if text[begin:begin+i] in dic:
            return text[begin:begin+i],i
    dic.add(text[begin])
    return text[begin],i#返回词以及词语的长度
def forward_segment_plus(text,dic):
    m = max_word_length #m赋值为最大词语长度
    word_list = [] #用于储存最终输出结果
    p_i = 0 #指针pi
    n = len(text)-p_i #表示指针pi到字符串末尾的长度
    while n >= 1: #指针没有扫描完时执行循环
        if m>n :
            m = n #如果词典中的最长字符长度大于字符到文本末尾的长度，就把最长字符长度设为字符到文本末尾的长度
        word1,i_plus = mini_segment(text,p_i,dic,m)#word1是在词典中的词或单字，i_plus是这个词的长度
        word_list.append(word1)
        p_i += i_plus
        n = len(text)-p_i #再次计算指针pi到字符串末尾的长度
    return word_list
```
