```python
import jieba
import sys,os
import math
import re
import numpy as np
from collections import defaultdict
```

```python
from nltk.probability import FreqDist  

f = open("in_the_name_of_people_segment.txt",encoding="utf-8")
text = f.read().split(" ")
 
stop_words = ['','\ufeff','©','_','—','【','】',')','(','、','，','“','”','。','\n','《','》',' ','-','！','？','.','\'','[',']','：','/','.','"','\u3000','’','．',',','…','?']
# 去停用词
text_list = [word for word in text if word not in stop_words]

# 词语最低熵
min_entropy = 0.8 
# 词语最低互信息
min_p = 7
# 词语最长长度
max_gram = 4
# 词语最低频率
min_freq = 5

# 生成候选词
def ngram(text_list,max_gram):
    unigram = [i for i in text_list]
    loop = len(unigram) + 1 - max_gram
    ngram = []
    for i in range(loop):
        ngram.append("_".join(text_list[i:i + max_gram]))
    if max_gram == 1:
        return unigram
    else:
        return ngram

    #计算词的概率
def probability(word):
    len_word = len(word.split("_"))
    total_count = len(word_all[len_word])
    prob = freq_all[len_word][word]/total_count
    return prob

def entropy(alist):
    f = FreqDist(alist)
    # 计算熵
#     ent = (-1)*sum([i/len(alist)*math.log(i/len(alist)) for i in f.values()])
    ent = 0
    for value in f.values():
        p = value/len(alist)
        ent += -1*p*math.log(p)
    return ent
 
def context(word):
    len_word = len(word.split("_"))
    indecies = np.where(np.array(word_all[len_word])==word)
    left_context = [word_all[len_word][i-1].split("_")[0] for i in indecies[0] if i > 0]
    right_context = [word_all[len_word][i+1].split("_")[-1] for i in indecies[0] if i < len(word_all[len_word])-1 ]
    return left_context,right_context

freq_all = [0]
word_all = [0]
for i in range(1,max_gram+1):
    t = ngram(text_list,i)
    freq = FreqDist(t)
    word_all.append(t)
    freq_all.append(freq)
    
#筛选一部分符合互信息的单词
final_word=[]
for i in range(2,max_gram+1):
    for j in word_all[i]:
        # 小于最小频率的直接过滤
        if freq_all[i][j] < min_freq:
            continue

        j_list = j.split("_")
        # 两个词以上的词，看成是两部分构成的，计算所有可能的组合的互信息，取最小的作为整个词的互信息
        p = min([probability("_".join(j_list[:k]))*probability("_".join(j_list[k:])) for k in range(1,len(j_list))])
        mi =  math.log(probability(j)/p)
        if mi > min_p:
            final_word.append((j,mi)) 
            
#         j_list = j.split("_")  
#         p_list = []
#         for k in range(1,len(j_list)):
#             #计算互信息
#             p = probability("_".join(j_list[:k]))*probability("_".join(j_list[k:]))
#             p_list.append(p)
#         p_min = min(p_list)
#         mi =  math.log(probability(j)/p_min)
#         if mi > min_p:
#             final_word.append((j,mi))
            
# 去重
final_word=list(set(final_word))

#筛选左右熵
final_result=[]
for word,mi in final_word:
    # 找出一个词前后出现的词
    left_context,right_context = context(word)
    left_entropy = entropy(left_context)
    right_entropy = entropy(right_context)
    # 取左右熵的最小值进行比较
    if min([right_entropy,left_entropy]) > min_entropy:
        final_result.append((word,mi,left_entropy, right_entropy))
        
print(final_result)
```

```python
# 按左右熵小的一个进行排序
list_sort = sorted(final_result,key=lambda x:min(x[2],x[3]),reverse=True)
```
